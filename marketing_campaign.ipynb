```python
df['Income'] = df['Income'].astype(int)
```

```python
## making a group of income levels ('low', 'mediam' , 'hight')
bins = [1000, 50000, 80000, 200000]
labels = ['low_income', 'mediam_income', 'hight_income']

df['Income_group'] = pd.cut(df['Income'], 
                              bins=bins, 
                              labels=labels,
                              include_lowest=True)
df.head(20)
```

```python
df['Income_group'].value_counts()
```

```python
## which one of income levels whos respons the most to the campaign
plt.style.use('fivethirtyeight')
Income_respons = df.groupby('Income_group')['Response'].sum()
label =df.groupby('Income_group').count()['ID'].index
plt.pie(Income_respons,labels=label,autopct='%1.3f%%',shadow=True,explode=[.05,.03,.03])
plt.title('Income Levels Rsponds To The Campaign')
```
